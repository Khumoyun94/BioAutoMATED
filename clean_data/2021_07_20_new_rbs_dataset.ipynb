```python
# from paper https://www.nature.com/articles/s41467-020-17222-4
# from github https://github.com/JeschekLab/uASPIre
# ml github here: https://github.com/BorgwardtLab/SAPIENs

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
seq_train = np.load('raw/rbs_sequences_train_validation.npy')
seq_train = list(seq_train)
print(len(seq_train))

out_train = np.load('raw/rbs_targets_train_validation.npy')
out_train = list(out_train)
print(len(out_train))

train = pd.DataFrame()
train['seq'] = seq_train
train['out'] = out_train
train.to_csv('clean/hollerer_rbs_train.csv', index = False)
train
```

```python
seq_test = np.load('raw/rbs_sequences_test.npy')
seq_test = list(seq_test)
print(len(seq_test))

out_test = np.load('raw/rbs_targets_test.npy')
out_test = list(out_test)
print(len(out_test))

test = pd.DataFrame()
test['seq'] = seq_test
test['out'] = out_test
test.to_csv('clean/hollerer_rbs_test.csv', index = False)
test
```
