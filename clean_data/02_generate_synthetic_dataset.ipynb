```python
import pandas as pd
import numpy as np
import random
import string
```

# Make small synthetic test set

```python
num_seqs = 1000
seq_len = 20
library = 'ATCG'

seqs = []
for i in range(num_seqs):
    seq = ''.join(random.choice(library) for i in range(seq_len))
    seqs.append(seq)
    
df = pd.DataFrame(seqs, columns = ['seq'])
df
```

# Helper functions for scoring

```python
def scoreStringPositive(seq):
    score = 0
    for s in seq:
        if s == 'A':
            add = 1
        elif s == 'T':
            add = 2
        elif s == 'C':
            add = 3
        elif s == 'G':
            add = 4
        score = score + add
    return score
```

```python
def scoreStringNegative(seq):
    score = 0
    for s in seq:
        if s == 'A':
            add = 1
        elif s == 'T':
            add = 2
        elif s == 'C':
            add = 3
        elif s == 'G':
            add = -1
        score = score + add
    return score
```

```python
df['positive_score'] = [scoreStringPositive(seq) for seq in df['seq']]
df['mixed_score'] = [scoreStringNegative(seq) for seq in df['seq']]
df.to_csv('clean/small_synthetic.csv', index = False)
df
```

# Large synthetic set

```python
# now do large synthetic set
num_seqs = 100000
seq_len = 20 # can change these
library = 'ATCG'

seqs = []
for i in range(num_seqs):
    seq = ''.join(random.choice(library) for i in range(seq_len))
    seqs.append(seq)
    
df = pd.DataFrame(seqs, columns = ['seq'])
df['positive_score'] = [scoreStringPositive(seq) for seq in df['seq']]
df['mixed_score'] = [scoreStringNegative(seq) for seq in df['seq']]
df.to_csv('clean/large_synthetic.csv', index = False)
df
```
