```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# folders of interest
file_name = 'results.txt'
folds = ['../outputs/autokeras/classification/', '../outputs/deepswarm/classification/', '../outputs/tpot/classification/']
mcc = pd.DataFrame()
for fold in folds:
    name = fold.split('/')[2]
    df = pd.read_csv(fold + file_name, sep = '\t', header = None)
    vals = df.values
    for i in range(len(vals)):
        val = vals[i]
        #print(val)
        if 'all values of MCC Class 0' in val[0]:
            new = val[0].split(": ")[1]
            nextline = vals[i+1][0]
            new = new + nextline
            new = new.strip("[")
            new = new.strip("]")

            new = [s for s in new.split(' ')]
            new = [s for s in new if s != '']
            new = [0 if s == '0.' else s for s in new]
            new = [float(s) for s in new]
            mcc[name] = new
mcc
```

```python
plt.figure(figsize=(3,3), dpi=300)
my_pal = {"autokeras": "darkturquoise", "deepswarm": "royalblue", 'tpot': 'teal'}
sns.barplot(data=mcc, edgecolor = 'black', palette = my_pal)
sns.swarmplot(data=mcc, edgecolor='black', palette = my_pal, linewidth = 1)
plt.xlabel('Architecture Search')
plt.xticks(rotation = 60)
plt.ylabel('MCC')
plt.tight_layout()
#plt.savefig('../out/ml_eval/1.png')
plt.show()
```
